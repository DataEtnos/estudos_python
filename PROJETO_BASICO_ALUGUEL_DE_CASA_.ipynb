```python
# Com .drop faço a exclusão de algumas colunas do data set

aluguel_casa = aluguel_casa.drop(["hoa","fire insurance", "furniture"],axis=1)
```

```python
#visualizo novamente o data set
aluguel_casa
```

```python
# Aqui faço o renomeamento das colunas para minha linguagem
aluguel_casa = aluguel_casa.rename(columns={"floor":"andar","index":"id","parking spaces":"vagas_de estacionamento","bathroom":"Banheiro","rent amount":"valor de aluguel","rooms":"Quartos","property tax":"contribuição predial","city":"cidade"})
```

```python
# Aqui fiz uma visulização de como ficou o dataset apos renomear
aluguel_casa
```

```python
# Aqui verifiquei que não tem dados nulos no dataset
aluguel_casa.isnull()
```

```python
# Esta é a função simplificada de Filtragem utilizo para ocupar menos espaço no notebook colabs alem de maximar meu tempo
# utilizo bem menos caracteres e tenho o mesmo resultado.

aluguel_casa =  aluguel_casa.query("total  == 4000" and  "Quartos == 3" and "animal == 'accept' " and "Banheiro == 1 or  Banheiro  == 2")
aluguel_casa.head()
```

```python
# Aqui fiz o agrupamento por cidade e preço ja dentro da expectativa que criei para saber a cidade mais barata .

agg_cidade_preco =  aluguel_casa.groupby(['cidade']) ["total"].agg("mean").reset_index()
agg_cidade_preco
```

```python
#Aqui fiz um agrupamento por qtd de quartos por cidade e tirando a media por total
agg_Quartos =  aluguel_casa.groupby(["Quartos" ,"cidade"])["total"].agg("mean").reset_index()
agg_Quartos
```

```python
#ultilizei a biblioteca seaborn para plotar um grafico do agrupamento que tirei do data set
#o grafico esta total por quartos !

ax =  sns.barplot(data = agg_Quartos , x = "Quartos", y = "total");
ax.set(xlabel = "Qtd de Quartos ", ylabel = "valor total por qtd de quartos");
```

```python
# Aqui fiz um agrupamento por andar , cidade e valor total pra comparar os valores em cada cidade e tirar as medias de  valor por cidade
agg_andar = aluguel_casa.groupby(["andar","cidade"])["total"].agg("mean").reset_index()
agg_andar
```

```python
 # Aqui fiz um grafico com o agrupamente por cidade e valor de moradia , para saber em qual cidade o valor é mais baixo .
 ax = sns.barplot(data= agg_andar,x="cidade",y="total");
 ax.set(xlabel="Cidades", ylabel=" Valor de Moradia");
```

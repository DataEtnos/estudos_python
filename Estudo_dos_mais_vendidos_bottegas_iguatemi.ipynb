```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.io as pio
import plotly.express as px
```

```python
# Agrupar por 'PRODUTO', 'CATEGORIA' e 'SKU' e somar os valores de 'TOTAL' e 'QTDE'
df_soma = df.groupby(['PRODUTO', 'CATEGORIA', 'SKU']).agg({'TOTAL': 'sum', 'QTDE': 'sum'}).reset_index()

# Ordenação
df_soma = df_soma.sort_values(by='TOTAL', ascending=False).reset_index(drop=True)

#FORMATAÇÃO
df_soma['TOTAL'] = df_soma['TOTAL'].apply(lambda x: f"R$ {x:,.2f}")
df_soma['QTDE'] = df_soma['QTDE'].apply(lambda x: f"{x}")

# Agrupando por 'PRODUTO', 'CATEGORIA' e 'SKU' e somar os valores de 'TOTAL' e 'QTDE'
df_soma = df.groupby(['PRODUTO', 'CATEGORIA', 'SKU']).agg({'TOTAL': 'sum', 'QTDE': 'sum'}).reset_index()

# Ordenar os resultados pelo valor total
df_soma = df_soma.sort_values(by='TOTAL', ascending=False).reset_index(drop=True)

# Formatar as colunas 'TOTAL' e 'QTDE' em reais
df_soma['TOTAL'] = df_soma['TOTAL'].apply(lambda x: f"R$ {x:,.2f}")
df_soma['QTDE'] = df_soma['QTDE'].apply(lambda x: f"{x}")

df_soma.head(30)
```

```python
df_vinho= df_soma[df_soma['CATEGORIA'] == 'VINHOS '][['SKU', 'PRODUTO', 'QTDE','TOTAL']].reset_index()

df_vinho.head(50)
```

```python
df_soma.info()
```

```python
df_soma['TOTAL'] = df_soma['TOTAL'].str.replace(r'\D', '', regex=True).astype(float)
```

```python
#top 20 produtos geral
top_20_produtos_mat = df_soma.nlargest(20, 'TOTAL')


```

```python
df_vinho['TOTAL'] = df_vinho['TOTAL'].str.replace(r'\D', '', regex=True).astype(float)
```

```python
#top 20 vinhos
top_20_produtos = df_vinho.nlargest(20, 'TOTAL')


```

```python
# Criar figura de barras
fig = px.bar(top_20_produtos, x='PRODUTO', y='TOTAL', color='PRODUTO',
             labels={'TOTAL': 'Total', 'PRODUTO': 'Produto'},
             title='Top 20 Produtos por vinho - iGUATEMI')

# Rotacionar rótulos do eixo x
fig.update_layout(xaxis=dict(tickangle=-45))

# Mostrar o gráfico
fig.show()


# Salvar o gráfico como um arquivo HTML
pio.write_html(fig, 'VINHO IGUATEMI.html', auto_open=True)



```

```python
# Criar gráfico de barras
fig = px.bar(top_20_produtos_mat, x='PRODUTO', y='TOTAL', color='PRODUTO',
             labels={'TOTAL': 'Total', 'PRODUTO': 'Produto'},
             title='Top 20 Produtos por Total')

# Rotacionar rótulos do eixo x
fig.update_layout(xaxis=dict(tickangle=-45))

# Mostrar o gráfico
fig.show()
```

```python
df_durabilidade= df.query('Durability > 300')[['Category', 'Name','Durability']]
```

```python
df_durabilidade
```

```python
# As 3 maiores armas com maior durabilidade do jogo
sns.barplot(data= df_durabilidade, x= "Name" , y = "Durability");
plt.figure(figsize=(300, 100))
```

```python
df
```

```python
df_boss_arm = df[df['Reinforcement'] == 'Boss'][['Category', 'Name', 'Durability']]
df_boss_arm
```

```python
#agrupando dados para saber as variação de durabilidade media de uma arma de boss do Darksoul 3
durability_by_category = df_boss_arm.groupby('Category')['Durability'].mean().reset_index()

# Crriando gráfico de barras horizontal
plt.figure(figsize=(10, 6))
sns.barplot(x='Durability', y='Category', data=durability_by_category, palette='viridis')
plt.xlabel('Durability')
plt.ylabel('Category')
plt.title('Average Durability of Boss Reinforcements by Category')
plt.show()
```
